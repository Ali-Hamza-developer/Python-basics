# Tuples

```python
my_tuple = ('apple', 'banana', 'cherry')
print(my_tuple[0])      # Output: 'apple'
print(my_tuple[-1])     # Output: 'cherry'
```

```python
my_tuple = (1, 2, 2, 3)

print(my_tuple.count(2))  # Output: 2
print(my_tuple.index(3))  # Output: 3
```

```python
# Packing
person = ("Ali", 25, "Pakistan")

# Unpacking
name, age, country = person
print(name)     # Ali
print(age)      # 25
```

```python
# -----------------------------
# üìò Basic Tuple Creation
# -----------------------------
print("Basic Tuple Creation:")
tuple1 = (1, 2, 3, 4)
tuple2 = "apple", "banana", "cherry"  # Without parentheses
tuple3 = (5,)  # Single-element tuple

print("Tuple 1:", tuple1)
print("Tuple 2:", tuple2)
print("Tuple 3 (one item):", tuple3)
print()

# -----------------------------
# üéØ Accessing Elements by Index
# -----------------------------
print("Accessing Tuple Elements:")
print("First item:", tuple1[0])
print("Last item:", tuple1[-1])
print()

# -----------------------------
# ‚úÇÔ∏è Slicing Tuples
# -----------------------------
print("Slicing Tuples:")
print("tuple1[1:3]:", tuple1[1:3])
print()

# -----------------------------
# üîÑ Looping Through a Tuple
# -----------------------------
print("Looping Through Tuple:")
for item in tuple2:
    print(item)
print()

# -----------------------------
# üß™ Tuple Methods
# -----------------------------
print("Tuple Methods:")
tuple4 = (1, 2, 2, 3, 4, 2)
print("Count of 2:", tuple4.count(2))
print("Index of 3:", tuple4.index(3))
print()

# -----------------------------
# üîÅ Tuple Packing & Unpacking
# -----------------------------
print("Tuple Packing and Unpacking:")
person = ("Ali", 25, "Pakistan")
name, age, country = person
print(f"Name: {name}, Age: {age}, Country: {country}")
print()

# -----------------------------
# üß± Nested Tuples
# -----------------------------
print("Nested Tuples:")
nested = ((1, 2), (3, 4), (5, 6))
print("Element at [1][0]:", nested[1][0])
print()

# -----------------------------
# ‚ùå Tuple is Immutable
# -----------------------------
print("Immutability Test:")
immutable_test = (10, 20, 30)
try:
    # This will raise an error
    immutable_test[0] = 100
except TypeError as e:
    print("Error:", e)
print()

# -----------------------------
# üîÅ Converting Between Tuple and List
# -----------------------------
print("Tuple <-> List Conversion:")
my_list = list(tuple1)
print("Converted to list:", my_list)

my_list[0] = 99
tuple1 = tuple(my_list)
print("Modified and back to tuple:", tuple1)
print()

# -----------------------------
# üìè Length of Tuple
# -----------------------------
print("Length of Tuple:")
print("Length of tuple2:", len(tuple2))
print()

# -----------------------------
# üí° Using Tuples as Dictionary Keys
# -----------------------------
print("Tuples as Dictionary Keys:")
locations = {
    (33.6844, 73.0479): "Islamabad",
    (24.8607, 67.0011): "Karachi"
}
print("City at (33.6844, 73.0479):", locations[(33.6844, 73.0479)])
print()

# -----------------------------
# üß† Tuple with Mixed Data Types
# -----------------------------
print("Tuple with Mixed Types:")
mixed = ("Python", 3.10, True, None)
for value in mixed:
    print(f"{value} ({type(value)})")
print()

# -----------------------------
# üìö Summary Comparison: Tuple vs List
# -----------------------------
print("Tuple vs List Comparison:")

my_list = [1, 2, 3]
my_tuple = (1, 2, 3)

print("List:", my_list)
print("Tuple:", my_tuple)

# Modifying list (OK)
my_list[0] = 99
print("Modified List:", my_list)

# Trying to modify tuple (Error)
try:
    my_tuple[0] = 99
except TypeError as e:
    print("Tuple Modification Error:", e)
```
